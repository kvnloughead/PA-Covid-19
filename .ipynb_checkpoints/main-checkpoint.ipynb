
# Pennsylvania Covid-19 Testing and New Cases Growth Rates

### An exploration of the relationship between reported rates growth in Covid-19 cases and the corresponding rates of growth in tests performed.

The last few days have seen stories in the news reporting an apparent slowing in the growth rates of new Covid-19 cases in Pennsylvania, and in the USA at large for that matter.  Officials have all levels of government have spoken of it _(find some links)_, and some have gone so far as to claim that _[paraphrasing, find quote]"we are no longer experiencing exponential growth"._   Of course, we have seen temporary dips in cases [before](cite Italy, maybe New York?), and most officials are careful to emphasize that this data is preliminary and uncertain and that we must therefore continue with our rigorous social distance and lockdown policies.  But I wondered whether these claims of reductions in growth rates, however qualified and tentative they may be, might not in fact be illusory:  whether they might not be a function of a reduction in _testing_, rather than of reduction in spread of the disease itself.

In order to see if my hypothesis held water, I wrote a Python program to scrape the [PA Coronavirus Update Archive](https://www.health.pa.gov/topics/disease/coronavirus/Pages/Archives.aspx), which keeps a daily record of the number of negative test results, positive test results, and deaths due to Covid-19 in Pennsylvania (among other data).  I couldn't find a proper dataset, so I wrote a program in Python to scrape the web page for the last couple of weeks of relevant data, which can be found in `parse_website.py` (imported as `p`).  Here is the compiled data.

```python
data = {'negatives': np.array(p.negative_tests), 
        'positives': np.array(p.positive_tests), 
        'deaths': np.array(p.deaths)}
df = pd.DataFrame(data)
df
```

One piece of critical information is the rate of positives, which can be calculated by

```python
rate_of_positive = data['positives']/(data['negatives'] + data['positives'])
df['positive_rate'] = rate_of_positive
df
```

The trend here is quite clear, if not particularly dramatic. We will also want to know the daily tally of new cases.  We could make the same calculation for new deaths each day, but for now we will focus on cases.  Note the appending of a single `np.nan` to fill out this column.   
The trend here is quite clear, if not particularly dramatic.   We will also want to know the daily tally of new cases.  We could make the same calculation for new deaths each day, but for now we will focus on cases. 


```python
new_cases = [p1 - p2 for p1, p2 in zip(df['positives'][:-1], df['positives'][1:])]
new_cases.append(np.nan)

df['new_cases'] = new_cases
df
```

Note the appending of a single `np.nan` to fill out this column.   Next, we compute the growth rate.


```python
nominal_growth_rate = [(p1 / p2) - 1 for p1, p2 in zip(df['positives'][:-1], df['positives'][1:])]
nominal_growth_rate.append(np.nan)

df['nominal_growth_rate'] = nominal_growth_rate
df
```

One can certainly see why people are saying the growth rate has gone down!  In fact there seem to be two distinct dips in the growth rate, one occuring at the end of March (roughly two weeks after drastic measures began to be taken in the first counties to be effected), and another one occuring just a few days ago.   However, let us run the same calculations for the daily number of tests being performed.

```python
total_tests = df['negatives'] + df['positives']

new_tests = [p1 - p2 for p1, p2 in zip(total_tests[:-1], total_tests[1:])]
new_tests.append(np.nan)

testing_growth_rate = [p1 / p2 for p1, p2 in zip(new_tests[:-1], total_tests[1:])]
testing_growth_rate.append(np.nan)

df['new_tests'], df['testing_growth_rate'] = [new_tests, testing_growth_rate]
df
```

Now the data is telling a rather different story, to my eyes.  There is a (not particulary surprising) connection between the number of tests performed and the number of new cases reported.  The earlier dip in the new case growth rate around the end of March comes hand in hand with a corresponding dip in the growth rate of the number of tests performed--as dip in growth rates that we've been experiencing lately.


Now, I am curious as to how closely these two numbers correspond. One way to find out would be to consider the how the rates of change themselves are changing. Thus

```python
cases_growth_rate_change = [(p / q) - 1 for p, q in zip(nominal_growth_rate[:-1], nominal_growth_rate[1:])]
cases_growth_rate_change.append(np.nan)

tests_growth_rate_change = [(p / q) - 1 for p, q in zip(testing_growth_rate[:-1], testing_growth_rate[1:])]
tests_growth_rate_change.append(np.nan)

df['cases_rate_change'] = cases_growth_rate_change
df['tests_rate_change'] = tests_growth_rate_change
```

```python
df2 = df[['negatives', 'positives', 'new_cases', 'new_tests', 
           'nominal_growth_rate', 'testing_growth_rate',
           'cases_rate_change', "tests_rate_change"]]

df2
```

```python
ax = plt.gca()

df.plot(kind='line', y='cases_rate_change', ax=ax)
df.plot(kind='line', y='tests_rate_change', color='red', ax=ax)

plt.show()
```

This seems to support my hypothesis:  if testing growth rate decreases or increases, so does the new case growth rate.

Next, we might wish to consider how many new cases would likely have been detected in the last 3 days had the daily increase in testing had _not_ dropped significantly.  To obtain a baseline testing growth rate, we will take the average of the rates from March 31 through April 5--a period of time with relatively static testing growth rates.  The average is thus

```python
mean_test_rate = df['testing_growth_rate'][3:10].mean()
mean_test_rate
```

And let's subset our dataframe, renaming some columns to save space.

```python
df_projected = {
    'neg': df['negatives'],
    'pos': df['positives'],
    'pos_rate': df['positive_rate'],
    'new_cases': df['new_cases'],
    'case_growth': df['nominal_growth_rate'],
    'new_tests': df['new_tests'],
    'testing_growth': df['testing_growth_rate']
}
df_projected = pd.DataFrame(df_projected)
df_projected
```

Now, lets update the three most recent days to reflect this average testing growth rate.  The easiest way to do this would be to take the number of new tests on the last day prior to the beginning of the most recent slowing of pace as the initial rate, and compute the projected tests per day for subsequent days from this.  Like so:

```python
new_tests_0 = df['new_tests'][4]
new_tests_projected = np.array([int(new_tests_0 * (1 + mean_test_rate)**i) for i in range(1, 5)][::-1])
new_tests_projected = np.append(new_tests_projected, df_projected['new_tests'][4:])
new_tests_projected

df_projected['new_tests_proj'] = new_tests_projected
df_projected['test_growth_proj'] = np.append(np.array([mean_test_rate]*4), df_projected['testing_growth'][4:])
df_projected
```

From here, we could project daily new cases by multiplying the projected daily tests by the rate of positivity.  

```python
new_cases_proj = df_projected['pos_rate'][:4] * df_projected['new_tests_proj'][:4]
new_cases_proj = np.array(new_cases_proj, dtype='int')
df_projected['new_cases_proj'] = np.append(new_cases_proj, df_projected['new_cases'][4:])
df_projected
```

```python
# total number of projected positives
pos_proj = [df_projected['pos'][4] + sum(new_cases_proj[:i+1]) for i in range(4)]
pos_proj = np.array(pos_proj[::-1], dtype='int')
df_projected['pos_proj'] = np.append(pos_proj, df_projected['pos'][4:])
df_projected
```

```python
ax = plt.gca()

df_projected.plot(kind='line', y='new_cases', ax=ax)
df_projected.plot(kind='line', y='new_cases_proj', color='red', ax=ax)
df_projected.plot(kind='line', y='new_tests', ax=ax)
df_projected.plot(kind='line', y='new_tests_proj', color='red', ax=ax)

plt.show()
```

The plot shows what is going well enough, I suppose.  If testing had continued to be ramped up as it had been doing, cases would be growing at essentially the same rate that they were.  Admittedly, the difference looks slight in this plot, but if the trend continues I expect it to grow more pronounced.

Now, the projection of cases per day is pretty clearly exponential.  Below you can see the projected growth rate in cases. It is curious, however, that the first day in my projections does in fact seem to drop significantly.  So perhaps there _was_ some sort of lessening in growth (or at least in perceived growth) _just prior_ to when testing rates dropped off.   Or perhaps something is wrong in my analysis?  It is also curious that my projected total cases doesn't actually turn out much higher than the reported total.  I have a hypothesis about that, which I will look at now.

We computed our projected new cases based on the reported rates of positive for each day being projected.  However, this rate of positivity is weighed down by weeks of less infection density.  Let's compute now each days rate of positivity for tests reported on that day.

```python
df_projected['daily_pos_rate'] = df_projected['new_cases'] / df_projected['new_tests']
df_projected
```

As you can see, the daily rate of positive tests is increasing much more quickly than was apparent from the overall increase in rates of positives.   Of course, some of this might be due to improved targetting of sick individuals, but I'm afraid I wouldn't know how to quantify that.  Let us recompute our projected new cases and total positives with these numbers.

```python
new_cases_proj_2 = df_projected['daily_pos_rate'][:4] * df_projected['new_tests_proj'][:4]
new_cases_proj_2 = np.array(new_cases_proj_2, dtype='int')
df_projected['new_cases_proj_2'] = np.append(new_cases_proj_2, df_projected['new_cases'][4:])
df_projected
```

Replotting:

```python
ax = plt.gca()

df_projected.plot(kind='line', y='new_cases', ax=ax)
df_projected.plot(kind='line', y='new_cases_proj_2', color='red', ax=ax)
df_projected.plot(kind='line', y='new_tests', ax=ax)
df_projected.plot(kind='line', y='new_tests_proj', color='red', ax=ax)

plt.show()
```
